```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
music_data = pd.read_csv('music.csv')
```

### Split the data
Divide the data into inputs and outputs to prepare for machine learning

```python
X = music_data.drop(columns=["genre"])
y = music_data['genre']
```

Initialize the model

```python
model = DecisionTreeClassifier()
```

Train the model

```python
model.fit(X.values, y)
```

Model Predictions

```python
model.predict([[21,1]])[0] # Predict what a 21 year old male would like
```

```python
model.predict([[21,0]])[0] # Predict what a 21 year old female would like
```

### Measure the accuracy of the model
Rule of thumb is to allocate 70-80% of the data to training and the remaining for testing
There needs to be enough training data (and sufficient fraction to test) to develop a well predicting model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) # Allocate 20% to testing
model = DecisionTreeClassifier() 
model.fit(X_train,y_train) # Fit the model
predictions = model.predict(X_test) # Make predictions based on the test set

score = accuracy_score(y_test, predictions)
score # Every time you run the score can be different
```

### Model Persistance
Saving trained models

```python
import joblib

joblib.dump(model, "music-recommender.joblib")
model = joblib.load('music-recommender.joblib')
```

### Visualizing Dicision trees

```python
from sklearn import tree

dec_tree = tree.plot_tree(model, 
                          feature_names=model.feature_names_in_, class_names=model.classes_, # Identify decider and result names
                          filled = True, rounded = True) # Color the tree and round the boxes
```

# Linear Regression

[Tutorial Link](https://www.youtube.com/playlist?list=PLzMcBGfZo4-mP7qA9cagf68V06sko5otr)


```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
data = pd.read_csv("students\student-mat.csv", sep=";")
data = data[["G1","G2","G3","studytime","failures","absences"]]

predict = "G3"
X = np.array(data.drop(columns =[predict]))
y = np.array(data[predict])

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2) # Split test-train data

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
acc = model.score(X_test, y_test)

print("Score of the model is: ", round(acc*100,2), " %") # Every time you run the score can be different
print("Co: ", model.coef_)
print("Intercept: ", model.intercept_)

predictions = model.predict(X_test)
```
